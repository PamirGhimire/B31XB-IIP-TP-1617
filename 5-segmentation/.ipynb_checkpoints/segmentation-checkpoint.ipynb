```python
%matplotlib inline
%pprint off

from __future__ import print_function, division

import numpy as np
import matplotlib.pyplot as plt

from skimage import img_as_ubyte
from skimage.color import rgb2gray
from skimage.io import imread, imshow
from skimage.measure import label, regionprops
from skimage.morphology import binary_closing, binary_opening, disk
from skimage.morphology import binary_dilation, binary_erosion
from skimage.filters import threshold_otsu
from skimage.transform import rescale

from skimage import feature
from skimage import measure, draw
```

## Segmentation of toy example

In this exercise, you will have to automatically find how many type of coins are present on the picture `coins.jpg`.
Therefore, read the image as `ubyte` and display it. In order to speed-up the subsequent processes, divide the scale of the image by a factor of 4.

```python
#read the coins image as a ubyte
coins_o = img_as_ubyte(imread('images/coins.jpg'))

#display original image
plt.figure()
plt.imshow(coins_o)
plt.title('coins image as-is')

#rescale coins
from skimage.transform import rescale
coins = rescale(coins_o, 0.25, order=1, mode='constant', cval=0, clip=True, preserve_range=False)

#display dimensions of original image and the rescaled image
print (coins_o.shape, coins.shape)
```

Subsquently, threshold the image using a global Otsu's thresholding after converting the image into gray scale. Display the image.

```python
#convert rgb coins to grey
coins_grey = rgb2gray(coins)

#compute otsu's threshold
from skimage import filters
otsu_coins = filters.threshold_otsu(coins)

#apply otsu's threshold to get binary image of coins
coins_binary_otsu = coins_grey < otsu_coins

#display binary image
plt.figure()
plt.imshow(coins_binary_otsu, cmap = 'Greys_r')
plt.title('binary image of coins computed using otsu_s threshold')
```

Clean the image using some morphological operations, i.e. erosion, dilation, opening, closing ...

```python
#clean binary image using closing; not great; dilate ; good
coins_cleaned = coins_binary_otsu
coins_cleaned = binary_closing(coins_cleaned)

nerosions = 3
for i in range(0, nerosions):
    coins_cleaned = binary_erosion(coins_cleaned)

ndilations = nerosions
for i in range(0, ndilations):
    coins_cleaned = binary_dilation(coins_cleaned)

#display cleaned image
plt.figure()
plt.imshow(coins_cleaned, cmap= 'Greys_r')
plt.title('otsu thresholded coins cleaned using dilation')
```

Using the functions [`label`](http://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.label) and [`regionprops`](http://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.regionprops) determine, the radius of each region and determine the number of type of coins in the image.

1. Show the image with the different labels.
1. How many regions do you find?
1. Display the radius of each region. Draw the corresponding circle on the image.

```python
#compute labels using 1-connectivity and 2-connectivity
labeled_coins_connected1, nregions_connected1 = label(coins_cleaned, neighbors=None, background=None, return_num=True, connectivity=1)
labeled_coins_connected2, nregions_connected2 = label(coins_cleaned, neighbors=None, background=None, return_num=True, connectivity=2)


print ('# of 1-connected regions = ', nregions_connected1)
print ('# of 2-connected regions = ', nregions_connected2)

#display labeled coins, labeled with regions of connectivity 1 and 2
plt.figure()
plt.imshow(labeled_coins_connected1, cmap = 'Greys_r')
plt.title('regions with connectivity 1')

plt.figure()
plt.imshow(labeled_coins_connected2, cmap = 'Greys_r')
plt.title('regions with connectivity 2')

#use region props to make measurements on detected regions

image = labeled_coins_connected1
regions = measure.regionprops(image)

forShow = coins_cleaned.astype('uint8')
mask = np.ones(forShow.shape)

for i in range(0, len(regions)):
    region = regions[i]
    
    c = region.centroid
    perimeter = region.perimeter
    r = perimeter / (2.0 * np.pi)
    
    rr, cc = draw.ellipse_perimeter(int(c[0]), int(c[1]), int(r), int(r))
    rr = rr.astype(int)
    cc = cc.astype(int)
    
    #manipulate image to plot circle
    for row in rr:
        for col in cc:
            mask[row][col] = 0

plt.figure()
plt.imshow(mask + forShow, cmap = 'Greys_r')
plt.title('detected coins with rectangular boundaries around them')
   
```

## Segmentation of markers

Considering the techniques you used in the previous exercise, try to segment different objects in `objets4.jpg` image.

Using segmentation find the number of each object in the image. 

Considering that this is not as straight forward as the previous example, come up with your creative solutions and in case of failar explain the reasons and problems. 

```python
#load objects4.jpg and show grayscale of image as is
objects_rgb = imread('images/objects4.jpg')
objects = rgb2gray(objects_rgb)

#show objects as-is
plt.figure()
plt.imshow(objects, cmap = 'Greys_r')
plt.title('objects4.jpg in grayscale')
```

```python
#idea
#use region-props to find the number of connected regions in the image (10 objects)
#measure the lengths (greater of the two dimensions) of the regions
#find the number of 'categories/classes'of lengths

#compute otsu's threshold to binarize the image
otsu_objects = threshold_otsu(objects)
objects_binary = objects <= otsu_objects

#display otsu_thresholded image 
plt.figure()
plt.imshow(objects_binary, cmap = 'Greys_r')
plt.title('otsu thresholded grayscale of objects4.jpg')

#add edge image to otsu-thresholded image and normalize
objects_edge = feature.canny(objects, sigma = 1)
plt.figure()
plt.imshow(objects_edge, cmap = 'Greys_r')
plt.title('edges in the grayscale')

#merge binary and edge images
objects_binary_edges = objects_binary + objects_edge
objects_binary_edges_n = objects_binary_edges / np.max(objects_binary_edges)

plt.figure()
plt.imshow(objects_binary_edges_n.astype('uint8'), cmap= 'Greys_r')
plt.title('thresholded image + edges')
```

```python
#perform opening = dilation, erosion
#attempt to fill the objects 

objects_filled = objects_binary_edges_n.astype('uint8')

#define custom opening
ndilations = 8
for i in range(0, ndilations):
    objects_filled = binary_dilation(objects_filled)
for i in range(0, ndilations):
    objects_filled = binary_erosion(objects_filled)
    
#erode once to get rid of corner-artifacts
objects_filled = binary_erosion(objects_filled)

#display results of custom-opening
plt.figure()
plt.imshow(objects_filled, cmap = 'Greys_r')
plt.title('regions filled with morphological operations')
```

```python
#use region-props to compute number of 1-conected regions
#compute labels using 1-connectivity and 2-connectivity
objects_connected1, nobjects_connected1 = label(objects_filled, neighbors=None, background=None, return_num=True, connectivity=1)
objects_connected2, nobjects_connected2 = label(objects_filled, neighbors=None, background=None, return_num=True, connectivity=2)

#display the number of regions discovered
print('number of 1-connected regions = ', nobjects_connected1)
print('number of 2-connected regions = ', nobjects_connected2)

#make measurements on detected regions
objects_regions = measure.regionprops(objects_connected1)

#record major-axis length of regions
lengths = np.zeros(len(objects_regions))

for i in range(0, len(objects_regions)):
    objecti = objects_regions[i]
    lengths[i] = objecti.major_axis_length


#find the number of classes/categories of lengths     
#we will say that a new category of length has been discovered when the current length 
#changes by >threshold% (say >10%) when traversing the sorted array of lengths
lengths = np.sort(lengths)
threshold = 0.1
ncategories = 1

for i in range(1, len(lengths)):
    prevlen = lengths[i-1]
    currlen = lengths[i]
    lenDeviation = np.absolute( (currlen - prevlen) / prevlen)
    
    if (lenDeviation > threshold):
        ncategories = ncategories + 1

print ('number of categories of lengths discovered = ', ncategories)
    
```
